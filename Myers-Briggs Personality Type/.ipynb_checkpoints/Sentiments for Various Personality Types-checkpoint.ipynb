# Can we predict personality type based on enthusiasm?

### (MBTI) Myers-Briggs Personality Type Dataset
Includes a large number of people's MBTI type and content written by them.

More about the MBTI: http://www.myersbriggs.org/my-mbti-personality-type/mbti-basics/home.htm?bhcp=1

Description of the dataset on Kaggle: https://www.kaggle.com/datasnaek/mbti-type

### Goal of this analysis
For this analysis I'm interested in whether people with different personality types express themselves differently, for example whether one personality type talks more about positive expressions, while another personality type is more neutral in the way they express themselves. 

```python
import pandas as pd
import numpy as np
import re
import nltk
from nltk import word_tokenize
from nltk import pos_tag
from nltk.sentiment.vader import SentimentIntensityAnalyzer
%matplotlib inline
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_style("darkgrid")

pd.set_option('display.max_colwidth', -1)
pd.set_option("display.max_rows", None)
pd.set_option('display.max_columns', None)
```

#### Clean dataset

```python
# remove urls starting with http 

# http = r'http\S+'
# P['clean_posts'] = P['posts'].str.replace(http, 'LINK' )

def preprocess(text):
    # remove urls starting with http and replace with a single word 
    text = re.sub(r'http\S+', 'LINK', text)
    # convert bytes to unicode utf-8
    text = unicode(text, 'utf8')
    # convert to lowercase
    text = text.lower()
    return text

P['clean_posts'] = P['posts'].apply(lambda x: preprocess(x))
```

### Perform sentiment analysis on the text

Sources:

- http://nlpforhackers.io/sentiment-analysis-intro/
- https://pythonprogramming.net/sentiment-analysis-module-nltk-tutorial/
- http://t-redactyl.io/blog/2017/04/using-vader-to-handle-sentiment-analysis-with-social-media-text.html

```python
sid = SentimentIntensityAnalyzer()
```

```python
def print_sentiment_scores(sentence):
    return sid.polarity_scores(sentence)
```

```python
# test
sentiment = print_sentiment_scores("I went for a walk today. The weather is beatifull")
print sentiment
print sentiment['compound']
```

#### Find the sentiment of each sentence that describes ones view of an event or attitude towards something

For this apply only the sentiment analyzer if the sentence contains a bigram of an adjective and a noun.

```python
# test
test = P.loc[2][2]

# split string into list of strings
test_sent = test.split('|||')
```

```python
# create a function that finds each pair of two sequential tags

def pairs(text, tag1='', tag2=''):
    tagged_pairs = nltk.bigrams(nltk.pos_tag(word_tokenize(text), tagset='universal'))
    pair = set([(a[0], b[0]) for (a, b) in tagged_pairs if ((a[1] == tag1) and 
                                                              (b[1] == tag2))])
    if (tag1 == 'ADJ') and (tag2 == 'NOUN'): 
        return [(adj, nn) for (adj, nn) in pair if any([adj.endswith('ing'), 
                                                        adj.endswith('ite'),
                                                        adj.endswith('ful'),
                                                        adj.endswith('cal'),
                                                        adj.endswith('y')])]
    elif (tag1 == 'ADV') and (tag2 == 'ADJ'):
        return [(adv, verb) for (adv, verb) in pair if any([(adv != 'so') and
                                                            (adv != 'how') and
                                                            (adv != 'not') and 
                                                            (adv != "n't")])]  
#     elif (tag1 == 'ADV') and (tag2 == 'ADV'):
#         return [(adv1, adv2) for (adv1, adv2) in pair if any([(adv1.endswith('ly'),
#                                                                adv2.endswith('ly'))])]
    else:
        return list(pair)
```

```python
# perform sentiment analysis if sentence contains a bigram of adjective and noun
for i in test_sent:
    if pairs(i, tag1='ADJ', tag2='NOUN'):
        print 'pair:', pairs(i, tag1='ADJ', tag2='NOUN')
        print i
        print 'pos: ', print_sentiment_scores(i)['pos'] 
        print 'neg: ', print_sentiment_scores(i)['neg'] 
        print '---'
```

```python
pos = []
for i in test_sent:
    if pairs(i, tag1='ADJ', tag2='NOUN'):
        pos.append(print_sentiment_scores(i)['pos'] )

# find the average of all the positive sentiments
print np.average(pos)

# find the median of all the positive sentiments
print np.median(pos)
```

```python
# show the variation in the predictions of positive sentiments
pos = []
for i in test_sent:
    if pairs(i, tag1='ADJ', tag2='NOUN'):
        pos.append(print_sentiment_scores(i)['pos'] )

print pos

fig, ax = plt.subplots(1,2, figsize=(8,4))
sns.boxplot(y=pos, orient='v', ax=ax[0], color="skyblue")
sns.distplot(pos, ax=ax[1])
plt.tight_layout()
```

```python
def PositiveSentimentScores(post, tag1="", tag2=""):
    
    # split post into list of strings
    post_sent = post.split('|||')
    pos = []
    for i in post_sent:
        if pairs(i, tag1=tag1, tag2=tag2):
            pos.append(print_sentiment_scores(i)['pos'] )
    return pos
```

#### Compare distributions of people with the same personality type and compare different personality types

```python
P['type'].unique()
```

```python
%%time
# add column for positive sentiment scores per post for sentences that describe an attitude towards a noun
P['pos_sent_scores_noun'] = P['clean_posts'].apply(lambda x: PositiveSentimentScores(x, 'ADJ', 'NOUN'))

# add column for positive sentiment scores per post for sentences that describe an attitude towards an adjective
P['pos_sent_scores_adj'] = P['clean_posts'].apply(lambda x: PositiveSentimentScores(x, 'ADV', 'ADJ'))
```

```python
INTP_pos_noun = P.loc[P['type']=='INTP']['pos_sent_scores_noun'].tolist()[:9]
INTP_pos_adj = P.loc[P['type']=='INTP']['pos_sent_scores_adj'].tolist()[:9]

fig,ax = plt.subplots(1,2, figsize=(15,7))
ax[0].boxplot(INTP_pos_noun)
ax[0].set_title("Positive sentiment variation for sentences that describe an attitude about a noun\n- INTP personality types -")

ax[1].boxplot(INTP_pos_adj)
ax[1].set_title("Positive sentiment variation for sentences that describe an attitude about an adjective\n- INTP personality types -");
```

```python
INTP_pos_noun = P.loc[P['type']=='INTP']['pos_sent_scores_noun'].tolist()[:9]
ENTP_pos_noun = P.loc[P['type']=='ENTP']['pos_sent_scores_noun'].tolist()[:9]

fig,ax = plt.subplots(1,2, figsize=(15,7))
ax[0].boxplot(INTP_pos_noun)
ax[0].set_title("Positive sentiment variation for sentences that describe an attitude about a noun\n- INTP personality types -")

ax[1].boxplot(ENTP_pos_noun)
ax[1].set_title("Positive sentiment variation for sentences that describe an attitude about a noun\n- ENTP personality types -");
```

#### Create new columns for the average of the postive scores and the most often occuring score

```python
P['avg_pos_sent_noun'] = P['pos_sent_scores_noun'].apply(lambda x: np.average(x))
P['median_pos_sent_noun'] = P['pos_sent_scores_noun'].apply(lambda x: np.median(x))

P['avg_pos_sent_adj'] = P['pos_sent_scores_adj'].apply(lambda x: np.average(x))
P['median_pos_sent_adj'] = P['pos_sent_scores_adj'].apply(lambda x: np.median(x))
```

```python
P.fillna(0, inplace=True)
```

### Can we predict personality type based on someones enthusiasm?

```python
from sklearn.svm import SVC, LinearSVC
from sklearn.cross_validation import train_test_split, cross_val_score, cross_val_predict, StratifiedKFold
from sklearn.metrics import confusion_matrix, classification_report
from sklearn.metrics import accuracy_score, recall_score, precision_score, average_precision_score, roc_auc_score, f1_score
from sklearn.grid_search import GridSearchCV
```

```python
# convert types to numeric classes
num_cat = {'INFJ':1, 'ENTP':2, 'INTP':3, 'INTJ':4, 'ENTJ':5, 'ENFJ':6, 'INFP':7, 'ENFP':8,
           'ISFP':9, 'ISTP':10, 'ISFJ':11, 'ISTJ':12, 'ESTP':13, 'ESFP':14}

P['type_num'] = P['type'].map(num_cat)
P.drop('type', axis=1, inplace=True)
```

```python
y = P.pop('type_num')
X = P[['avg_pos_sent_noun', 'median_pos_sent_noun', 'avg_pos_sent_adj', 'median_pos_sent_adj']]
```

```python
svm = SVC()
ypred = cross_val_predict(svm, X, y, cv=5, n_jobs=-1)
```

```python
print "accuracy:\t\t", accuracy_score(y, ypred)
```

#### Conclusion
Given this very low accuracy score we can not predict personality type based on enthusiasm in the style of writing.

### Tune parameters

```python
Xtrain, Xtest, ytrain, ytest = train_test_split(y, X, test_size=0.2)

print len(Xtrain)
print len(Xtest)
```

```python
parameters_svm = {
    'svm__kernel': ('linear', 'rbf'),
    'svm__C': (1, 10),
    'svm_shrinking': (True, False),
}

param_grid = [
  {'C': [1, 10], 'kernel': ['linear']},
  {'C': [1, 10], 'gamma': [0.001, 0.0001], 'kernel': ['rbf']},
 ]
```

```python
def svc_param_selection(X, y):
    Cs = [0.001, 1, 10]
    gammas = [0.001, 0.1]
    param_grid = {'C': Cs, 'gamma' : gammas}
    grid_search = GridSearchCV(svm, param_grid, cv=5)
    grid_search.fit(X, y)
    grid_search.best_params_
    return grid_search.best_params_
```

```python
svc_param_selection(Xtrain, ytrain)
```

```python
# instantiate the grid
grid_svm = GridSearchCV(svm, param_grid, cv=5, n_jobs=-1, verbose=1)
grid_svm.fit(Xtrain, ytrain)
```
