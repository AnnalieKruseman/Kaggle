#### Load libraries

```python
import pandas as pd
import numpy as np
%matplotlib inline
import matplotlib.pyplot as plt
import seaborn as sns
import json
```

#### Load datasets

```python
T = pd.read_csv('USvideos.csv')
```

```python
T.columns
```

```python
# map category to category_id
id_to_category = {}
with open('US_category_id.json', 'r') as f:
    data = json.load(f)
    for category in data['items']:
        id_to_category[category['id']] = category['snippet']['title']
        id_to_category = {int(str(k)): str(v) for k, v in id_to_category.items()}

T['category'] = T['category_id'].map(id_to_category)
```

```python
# create list of tags for each video
T['tags'] = T['tags'].apply(lambda x: x.split('|')).head()
```

## Descriptive Analysis

```python
T.head()
```

#### How many videos are there in each category?

```python
T['category'].value_counts().plot(kind='bar');
```

#### How many channels are there per category?

```python
T.groupby('category')['channel_title'].size().sort_values(ascending=False).plot(kind='bar');
```

#### How many views are there per category?

```python
T.groupby('category')['views'].sum().sort_values(ascending=False).plot(kind='bar');
```

#### How do likes versus dislikes vary per category?

If this ratio is larger than 1, this indicates that there are more positive reactions about this video than negative reactions. 

If this ratio is smaller than 1, this indicates that there are more negative reactions.

```python
# compute a ratio for like versus dislikes per category 
T['likes/dislikes'] = T['likes']/T['dislikes']
```

```python
print len(T)
print len(T.loc[T['likes/dislikes']<1])
print len(T.loc[T['likes/dislikes']>1])
```

```python
# create labels for a video indicating whether the majority of the comments contain likes or dislikes
labels = ['majority dislikes', 'majority likes']
bins = [0,1,np.inf]

# remove the missing values from the likes/dislikes columns
temp = T['likes/dislikes'].dropna()#replace(np.inf, np.nan).dropna()
T['majority likes/dislikes'] = pd.cut(temp, bins=bins, labels=labels)
```

```python
fig, ax = plt.subplots(figsize=(15,7))
T.groupby(['category', 'majority likes/dislikes']).size().unstack().plot(kind='bar', ax=ax)
plt.title("Majority of videos with likes versus dislikes per category")
plt.ylabel("Frequency");
```

```python
T.loc[T['likes/dislikes']<1]
```

#### How does the distribution of the ratio likes versus dislikes look like for each category?

```python
sns.set_style("whitegrid")

fig, ax = plt.subplots(figsize=(15,7))
ax = sns.boxplot(x="category", y="likes/dislikes", data=T.loc[T['likes/dislikes']<100], ax=ax)
plt.xticks(rotation=80);
```

By looking at this data, we see that overall people view YouTube videos because they like to watch videos that give them a good feeling. 

We also see that for the category News & Politics people show more dissatisfaction with a video.
With respect to all other categories we wouldn't be very suprised with these results.
